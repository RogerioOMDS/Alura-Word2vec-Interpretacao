```python
import string
import nltk
nltk.download('punkt')

def tokenizador(texto):
  texto = texto.lower() # deixa o texto minusculo
  lista_alfanumerico = [] 

  for token_valido in nltk.word_tokenize(texto): # para cada elemento do texto
    if token_valido in string.punctuation: continue # se o elemento for pontuação, ignora
    lista_alfanumerico.append(token_valido) # se for palavra, salva na lista

  return lista_alfanumerico
```

```python
palavras_teste = tokenizador("Hoje é um novo dia, de um novo tempo, que começou")
print(palavras_teste)
```

```python
string.punctuation
```

```python
# criando a função para tornar as frases em vetores 

import numpy as np
def combinacao_de_vetores_por_soma(palavras_numeros):
  vetor_resultante = np.zeros(300) # tamanho do modelo já treinado
  for pn in palavras_numeros:
    try: 
      vetor_resultante += modelo.get_vector(pn)
    except KeyError:
      if pn.isnumeric(): # segundo o artigo do nilc, todo os numeros devem ser convertidos para zeros
        pn = "0"*len(pn)
        vetor_resultante += modelo.get_vector(pn)
      else:
        vetor_resultante += modelo.get_vector('unknown') # também segundo o artigo, todas as palavras desconhecidas devem ser convertidas para 'unknown'
    
  return vetor_resultante
```

```python
palavras_numeros = tokenizador('Hoje, dia 14 do 16 onde o sol torce a meia noite na cidade de Vakulkex')
vetor_texto = combinacao_de_vetores_por_soma(palavras_numeros)
len(vetor_texto)
```

```python
# agora, fazer a vetorização de todos os títulos do texto
def matriz_vetores(textos):
  x = len(textos)
  y = 300
  matriz = np.zeros((x,y))

  for i in range(x):
    palavras_numeros = tokenizador(textos.iloc[i])
    matriz[i] = combinacao_de_vetores_por_soma(palavras_numeros)

  return matriz
```

```python
matriz_vetores_treino = matriz_vetores(artigo_treino.title)
matriz_vetores_teste = matriz_vetores(artigo_teste.title)
print(matriz_vetores_treino.shape)
print(matriz_vetores_teste.shape)
```

```python
from sklearn.linear_model import LogisticRegression
LR = LogisticRegression(max_iter=200)

LR.fit(matriz_vetores_treino, artigo_treino.category)
```

```python
LR.n_iter_ # quantas iterações foram necessárias para convergir
```

```python
LR.score(matriz_vetores_teste, artigo_teste.category)
```

```python
artigo_teste.category.unique() # Quais categorias temos 
```

```python
from sklearn.metrics import classification_report

label_prevista = LR.predict(matriz_vetores_teste)
CR = classification_report(artigo_teste.category, label_prevista)
print(CR)
```

```python
# 0.86 de precisão significa que apenas 86% do que estou classificando como coluna é de fato coluna 
# 0.72 de recall significa que o que classifiquei como coluna representa apenas 72% do que é coluna de fato. 
# macro avg é a média comum dos valores
# weighted avg é a média ponderada dos valores, onde os pesos são as quantidades de dados
```

```python
# Descompactando o SkipGram300
!unzip "/content/drive/MyDrive/Alura/Curso_word2vec/skip_s300.zip"
modelo_skipgram = KeyedVectors.load_word2vec_format("skip_s300.txt")
```

```python
def combinacao_de_vetores_por_soma_skipgram(palavras_numeros):
  vetor_resultante = np.zeros(300)
  for pn in palavras_numeros:
    try: 
      vetor_resultante += modelo_skipgram.get_vector(np)

    except KeyError:
      if pn.isnumeric():
        pn = "0"*len(pn)
        vetor_resultante += modelo_skipgram.get_vector(pn)
      else:
        vetor_resultante += modelo_skipgram.get_vector("unknown")

  return vetor_resultante
```

```python
def matriz_vetores_skipgram(textos):
  x = len(textos)
  y = 300
  matriz = np.zeros((x,y))

  for i in range(x):
    palabras_numeros = tokenizador(textos.iloc[i])
    matriz[i] = combinacao_de_vetores_por_soma_skipgram(palavras_numeros)

  return matriz
```

```python
matriz_vetores_treino_skipgram = matriz_vetores_skipgram(artigo_treino.title)
matriz_vetores_teste_skipgram = matriz_vetores_skipgram(artigo_teste.title)
```

```python
LR_skipgram = LogisticRegression(max_iter=1000)
LR_skipgram.fit(matriz_vetores_treino_skipgram, artigo_treino.category)
label_previsao_skipgram = LR_skipgram.predict(matriz_vetores_teste)

CR_skipgram = classification_report(artigo_teste.category, label_previsao_skipgram)
print(CR_skipgram)
```
